```python
import re

def detect_hinglish(text):
    if not isinstance(text, str):  # handle NaN or non-string
        return "unknown"

    has_hindi = bool(re.search(r'[\u0900-\u097F]', text))
    has_english = bool(re.search(r'[a-zA-Z]', text))

    words = set(text.lower().split())



    if has_hindi:
        return 'Hindi'
    elif has_english:
        return 'English'
    else:
        return 'Other/None'
```

```python
import re
import emoji
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from fuzzywuzzy import process
import pandas as pd
from tqdm import tqdm

# NLTK downloads
nltk.download("punkt")
nltk.download("wordnet")
nltk.download("stopwords")

lemmatizer = WordNetLemmatizer()
english_stopwords = set(stopwords.words("english"))

# Base abusive words dictionary
ABUSE_DICT = {
    "bc": "behenchod",
    "b.c": "behenchod",
    "b!c": "behenchod",
    "mc": "madarchod",
    "m.c": "madarchod",
    "c**t": "chutiya",
    "ch*tiya": "chutiya",
    "chu*ya": "chutiya",
    "bkl": "behen ke laude",
    "bhkl": "behen ke laude",
}

# Automatically generate common obfuscation variations
def generate_variations(word):
    variants = set()
    variants.add(word)
    variants.add(word.replace("c", "*"))
    variants.add(word.replace("c", "!"))
    variants.add(word.replace("o", "0"))
    variants.add(word.replace("e", "3"))
    return variants

for key in list(ABUSE_DICT.keys()):
    for v in generate_variations(key):
        ABUSE_DICT[v] = ABUSE_DICT[key]

# Hinglish pattern normalization
HINGLISH_PATTERNS = {
    r"g(a|aa|aaa|)ndu+": "gandu",
    r"m(a|aa|aaa|)d(a|aa)rch(o|oo)d": "madarchod",
    r"b(e|ee|i)h(e|i)nchod": "behenchod",
    r"ch(u|oo)t(iy|iya|iya*)": "chutiya",
}

def normalize_hinglish(word):
    for p, norm in HINGLISH_PATTERNS.items():
        if re.fullmatch(p, word):
            return norm
    return word

# Emoji conversion (dynamic)
def convert_emoji_auto(text):
    text = emoji.demojize(text)
    text = re.sub(r":([a-z_]+):", r" \1 ", text)  # replace all emojis with their name
    return text

# Deobfuscate symbols and numbers
def deobfuscate(text):
    # Remove everything except letters and spaces
    text = re.sub(r"[^a-zA-Z\s]", " ", text)

    # Collapse multiple spaces into one
    text = re.sub(r"\s+", " ", text).strip()
    return text

# Reduce elongated words
def reduce_lengthening(text):
    return re.sub(r"(.)\1{2,}", r"\1", text)  # aaa -> a

# Expand known abusive words / slangs
def expand_slang(text):
    words = text.split()
    new = []
    for w in words:
        lw = w.lower()
        if lw in ABUSE_DICT:
            new.append(ABUSE_DICT[lw])
        else:
            new.append(w)
    return " ".join(new)

# Lemmatize and remove stopwords
def normalize_english(word):
    w = lemmatizer.lemmatize(word)
    if w not in english_stopwords:
        return w
    return ""

# Fuzzy matching for unknown abusive words
def fuzzy_abuse_normalize(word, abuse_list, threshold=85):
    match, score = process.extractOne(word, abuse_list)
    if score >= threshold:
        return match
    return word

# Full preprocessing pipeline
def preprocess_text(text, use_fuzzy=True):
    text = text.lower()
    text = convert_emoji_auto(text)
    text = deobfuscate(text)
    text = reduce_lengthening(text)
    text = expand_slang(text)

    words = nltk.word_tokenize(text)
    normalized_words = []
    abuse_list = list(ABUSE_DICT.values())

    for w in words:
        w = normalize_hinglish(w)
        w = normalize_english(w)
        if not w:
            continue
        if use_fuzzy:
            w = fuzzy_abuse_normalize(w, abuse_list)
        normalized_words.append(w)

    return " ".join(normalized_words)
```
