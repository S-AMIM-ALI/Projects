```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_data = np.array([[0.5, 0.3, 0.8],
                       [0.1, 0.7, 0.2],
                       [0.4, 0.6, 0.9]])

# Example weights
weights = np.array([2, 6, -10])

# Bias
bias = 0.3
```

```python
input_data.shape
```

```python
input_data
```

```python
def sigmoid_activation(x):
    return 1 / (1 + np.exp(-x))
```

```python
def single_neuron(input_data, weights, bias):
    # Perform the weighted sum of inputs
    weighted_sum = np.dot(input_data.T, weights) + bias

    # Apply activation function
    output = sigmoid_activation(weighted_sum)

    return output
```

```python
outputs = single_neuron(input_data, weights, bias)

print(f"Inputs : {input_data}")
print(f"Outputs : {outputs}")
```

```python
plt.scatter(range(len(outputs)), outputs)
plt.plot(range(len(outputs)), outputs)
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Single Neuron Output')
plt.grid(True)
```
