```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples=200, n_features=1, n_targets=1, noise=20, random_state=40)
```

```python
plt.scatter(X, y, color = "red", marker = "x")
plt.xlabel("Independent Feature")
plt.ylabel("Dependent Feature")
plt.title("Scatter Plot")
plt.show
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
```

```python
y = scaler.fit_transform(y.reshape(-1, 1))
```

```python
plt.scatter(X, y, color = "orange", marker = "x")
plt.xlabel("Independent Feature")
plt.ylabel("Dependent Feature")
plt.title("Scatter Plot, After preprocessing Data.")
plt.show()
```

```python
sns.distplot(X, color = "red")
sns.distplot(y, color = "black")
plt.title("Distribution Plot")
plt.show
```

```python
theta0 = 0
theta1 = 0
```

```python
plt.scatter(X, y, color = "#FA5417", marker = "x")
plt.plot(X, theta0 + theta1*X, c = 'g')
plt.xlabel("Independent Feature")
plt.ylabel("Dependent Feature")
plt.title("Scatter Plot for Randomly theta0, theta1")
plt.show()
```

```python
eta = 0.01                                # learning rate
n = len(X)                                
for i in range(1000):                     
    y_pred = theta0 + theta1*X            
    dL0 = (-2/n)*np.sum(y - y_pred)       
    dL1 = (-2/n)*np.sum((y - y_pred)*X)   
    theta0 = theta0 - eta*dL0             
    theta1 = theta1 - eta*dL1  
```

```python
print(f"updated value of theta 0 : {theta0}")
print(f"updated value of theta 1 : {theta1}")
```

```python
plt.scatter(X, y, color = "#FA5417", marker = "x")
plt.plot(X, theta0 + theta1*X, c = 'b')
plt.xlabel("Independent Feature")
plt.ylabel("Dependent Feature")
plt.title("Scatter Plot, After applying Gradient Descent")
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 8))


eta = [0.1, 0.01, 0.001, 0.0001]
n = len(X)                                # total no. of data points
a = 1
for j, ax in enumerate(axes.flat):
    
    theta0 = 0
    theta1 = 0
    for i in range(1000):                     # iterating 1000 times
        y_pred = theta0 + theta1*X            # Calculate predicted value
        dL0 = (-2/n)*np.sum(y - y_pred)       # calculate the loss w.r.t theta0
        dL1 = (-2/n)*np.sum((y - y_pred)*X)   # Calculate the loss w.r.t theta1
        theta0 = theta0 - eta[j]*dL0             # update the theta0 value to minimize the loss
        theta1 = theta1 - eta[j]*dL1             # update the theta1 value to minimize the loss
    ax.scatter(X, y, color = "#FA5417", marker = "x")
    ax.plot(X, theta0 + theta1*X, c = 'b')
    ax.set_title(f"Scatter Plot : eta = {eta[j]}")
```
