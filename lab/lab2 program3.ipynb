```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
from sklearn.datasets import make_regression
X, y = make_regression(n_samples=100, n_features=3, n_informative=3, n_targets=1, noise = 0.005, random_state=42)
```

```python
plt.scatter(X[:, 0], y, color = "green", marker = "*", label = " Feature1")
plt.scatter(X[:, 1], y, color = "orange", marker = "x", label = "Feature2")
plt.scatter(X[:, 2], y, color = "blue", marker = "+", label = "Feature3")
plt.xlabel("Independent Feature")
plt.ylabel("Dependent Feature")
plt.legend(loc = 1)
plt.title("Scatter Plot")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
y = scaler.fit_transform(y.reshape(-1, 1))
```

```python
plt.scatter(X[:, 0], y, color = "green", marker = "*", label = " Feature1")
plt.scatter(X[:, 1], y, color = "orange", marker = "x", label = "Feature2")
plt.scatter(X[:, 2], y, color = "blue", marker = "+", label = "Feature3")
plt.xlabel("Independent Feature")
plt.ylabel("Dependent Feature")
plt.legend(loc = 1)
plt.title("Scatter Plot")
plt.show()
```

```python
theta0 = 0.0
theta1 = 0.0
theta2 = 0.0
theta3 = 0.0
```

```python
eta = 0.01                                # learning rate
n = len(X)                                # total no. of data points
for i in range(1000):                     # iterating 1000 times
    y_pred = theta0 + theta1*X[:, 0] + theta2*X[:, 1] + theta3*X[:, 2]            # Calculate predicted value
    dL0 = (-2/n)*np.sum(y - y_pred)       
    dL1 = (-2/n)*np.sum((y - y_pred)*X[:, 0])   
    dL2 = (-2/n)*np.sum((y - y_pred)*X[:, 1])   
    dL3 = (-2/n)*np.sum((y - y_pred)*X[:, 2])   
    theta0 = theta0 - eta*dL0            
    theta1 = theta1 - eta*dL1            
    theta2 = theta2 - eta*dL2            
    theta3 = theta3 - eta*dL3 
```

```python
print(f"updated value of theta 0 : {theta0}")
print(f"updated value of theta 1 : {theta1}")
print(f"updated value of theta 2 : {theta2}")
print(f"updated value of theta 3 : {theta3}")
```
