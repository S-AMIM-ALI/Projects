```python
import numpy as np

# Activation function (Step function) for Madaline
def step_function(x):
    return 1 if x >= 0 else -1

# XOR Training data
# Input pattern: (x1, x2), Target output: y
X = np.array([[1, 1], [1, -1], [-1, 1], [-1, -1]])  # Inputs for XOR
Y = np.array([-1, 1, 1, -1])  # Expected outputs for XOR

# Weights for the first layer (hidden layer)
# Let's say we use 2 hidden neurons, so we initialize two weights for each input and bias
w1 = np.random.uniform(-1, 1, 2)  # Weights for hidden layer neuron 1
w2 = np.random.uniform(-1, 1, 2)  # Weights for hidden layer neuron 2
b1 = np.random.uniform(-1, 1)  # Bias for hidden layer neuron 1
b2 = np.random.uniform(-1, 1)  # Bias for hidden layer neuron 2

# Weights for the output layer
w_out = np.random.uniform(-1, 1, 2)  # Weights for output layer neurons
b_out = np.random.uniform(-1, 1)  # Bias for output neuron

# Learning rate
learning_rate = 0.5

# Training the Madaline network (for simplicity, let's train it for a few iterations)
for epoch in range(10000):
    for i in range(len(X)):
        # Forward pass for hidden layer
        net1 = np.dot(X[i], w1) + b1
        net2 = np.dot(X[i], w2) + b2
        out1 = step_function(net1)  # Output from hidden neuron 1
        out2 = step_function(net2)  # Output from hidden neuron 2
        
        # Output layer computation (weighted sum of hidden layer outputs)
        net_out = np.dot([out1, out2], w_out) + b_out
        output = step_function(net_out)  # Final output from the network
        
        # Compute the error (difference between expected and actual output)
        error = Y[i] - output
        
        # Backpropagation (updating the weights)
        if error != 0:
            # Update weights for output layer
            w_out += learning_rate * error * np.array([out1, out2])
            b_out += learning_rate * error
            
            # Update weights for hidden layer 1
            w1 += learning_rate * error * w_out[0] * X[i]
            b1 += learning_rate * error * w_out[0]
            
            # Update weights for hidden layer 2
            w2 += learning_rate * error * w_out[1] * X[i]
            b2 += learning_rate * error * w_out[1]

# Test the network
print("Final weights after training:")
print(f"w1: {w1}, w2: {w2}, w_out: {w_out}, b1: {b1}, b2: {b2}, b_out: {b_out}")

print("\nPredictions after training:")
for i in range(len(X)):
    net1 = np.dot(X[i], w1) + b1
    net2 = np.dot(X[i], w2) + b2
    out1 = step_function(net1)
    out2 = step_function(net2)
    net_out = np.dot([out1, out2], w_out) + b_out
    output = step_function(net_out)
    print(f"Input: {X[i]} | Predicted Output: {output} | Expected Output: {Y[i]}")
```
