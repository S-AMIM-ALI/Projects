# Multiple Linear Regression using Gradient Descent

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
from sklearn.datasets import make_regression
X, y = make_regression(n_samples=100, n_features=3, n_informative=3, n_targets=1, noise = 0.005, random_state=42)
```

# 1) Use matplotlib to perform data analysis on the dataset and plot scatter of each feature vs target, plot distribution of each feature and target

```python
plt.scatter(X[:, 0], y, color = "#DA3048", marker = "*", label = "1 Feature")
plt.scatter(X[:, 1], y, color = "orange", marker = "x", label = "2 Feature")
plt.scatter(X[:, 2], y, color = "pink", marker = "+", label = "3 Feature")
plt.xlabel("Independent Feature")
plt.ylabel("Dependent Feature")
plt.legend(loc = 1)
plt.title("Scatter Plot")
plt.show()
```

# 2) Normalize the dataset

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
y = scaler.fit_transform(y.reshape(-1, 1))
```

```python
plt.scatter(X[:, 0], y, color = "#DA3048", marker = "*", label = "1 Feature")
plt.scatter(X[:, 1], y, color = "orange", marker = "x", label = "2 Feature")
plt.scatter(X[:, 2], y, color = "pink", marker = "+", label = "3 Feature")
plt.xlabel("Independent Feature")
plt.ylabel("Dependent Feature")
plt.legend(loc = 1)
plt.title("Scatter Plot")
plt.show()
```

```python
# let assume Initial random value of theta0, theta1
theta0 = 0.0
theta1 = 0.0
theta2 = 0.0
theta3 = 0.0
```

# 3) Apply gradient descent

```python
# Gradien Descent iterating 1000 times to fit the curve.
eta = 0.01                                # learning rate
n = len(X)                                # total no. of data points
for i in range(1000):                     # iterating 1000 times
    y_pred = theta0 + theta1*X[:, 0] + theta2*X[:, 1] + theta3*X[:, 2]            # Calculate predicted value
    dL0 = (-2/n)*np.sum(y - y_pred)       # calculate the loss w.r.t theta0
    dL1 = (-2/n)*np.sum((y - y_pred)*X[:, 0])   # Calculate the loss w.r.t theta1
    dL2 = (-2/n)*np.sum((y - y_pred)*X[:, 1])   # Calculate the loss w.r.t theta2
    dL3 = (-2/n)*np.sum((y - y_pred)*X[:, 2])   # Calculate the loss w.r.t theta3
    theta0 = theta0 - eta*dL0             # update the theta0 value to minimize the loss
    theta1 = theta1 - eta*dL1             # update the theta1 value to minimize the loss
    theta2 = theta2 - eta*dL2             # update the theta2 value to minimize the loss
    theta3 = theta3 - eta*dL3             # update the theta3 value to minimize the loss
```

# 4) Calculate the coefficients

```python
print(f"updated value of theta 0 : {theta0}")
print(f"updated value of theta 1 : {theta1}")
print(f"updated value of theta 2 : {theta2}")
print(f"updated value of theta 3 : {theta3}")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 8))

eta = [0.1, 0.01, 0.001, 0.0001]
n = len(X)                                # total no. of data points
for j, ax in enumerate(axes.flat):
    
    theta0 = 0
    theta1 = 0
    theta2 = 0.0
    theta3 = 0.0
    for i in range(1000):                     # iterating 1000 times
        y_pred = theta0 + theta1*X[:, 0] + theta2*X[:, 1] + theta3*X[:, 2]            # Calculate predicted value
        dL0 = (-2/n)*np.sum(y - y_pred)       # calculate the loss w.r.t theta0
        dL1 = (-2/n)*np.sum((y - y_pred)*X[:, 0])   # Calculate the loss w.r.t theta1
        dL2 = (-2/n)*np.sum((y - y_pred)*X[:, 1])   # Calculate the loss w.r.t theta2
        dL3 = (-2/n)*np.sum((y - y_pred)*X[:, 2])   # Calculate the loss w.r.t theta3
        theta0 = theta0 - eta[j]*dL0             # update the theta0 value to minimize the loss
        theta1 = theta1 - eta[j]*dL1             # update the theta1 value to minimize the loss
        theta2 = theta2 - eta[j]*dL2             # update the theta2 value to minimize the loss
        theta3 = theta3 - eta[j]*dL3             # update the theta3 value to minimize the loss
    ax.scatter(X[:, 0], y, color = "#DA3048", marker = "*", label = "1 Feature")
    ax.scatter(X[:, 1], y, color = "orange", marker = "x", label = "2 Feature")
    ax.scatter(X[:, 2], y, color = "pink", marker = "+", label = "3 Feature")
    ax.plot(X, theta0 + theta1*X[:, 0] + theta2*X[:, 1] + theta3*X[:, 2], c = 'r')
    ax.set_title(f"Scatter Plot : eta = {eta[j]}")
    print(f"Coefficients and Intercepts : eta = {eta[j]}")
    print(f"updated value of theta 0 : {theta0}")
    print(f"updated value of theta 1 : {theta1}")
    print(f"updated value of theta 2 : {theta2}")
    print(f"updated value of theta 3 : {theta3}")
    print()
```
