#Point-wise SGD

```python
def sgd(X, y, learning_rate=0.1, num_iterations=1000):

    MSE_list = []
    R2_list = []
    param_0_list = []
    param_1_list = []

    m = len(X)
    param_0 = 0
    param_1 = 0

    for epoch in range(num_iterations):

      for i in range(m):

          y_pred = param_1 * X[i] + param_0
          loss = y_pred - y[i]

          grad_0 = loss * 2
          grad_1 = loss * X[i] * 2

          param_0 -= learning_rate * grad_0
          param_1 -= learning_rate * grad_1

      y_bar = np.mean(y)
      y_pred_all = param_1 * X + param_0

      loss_all = y_pred_all - y
      total = y_bar - y

      MSE = np.mean(loss_all ** 2)
      MST = np.mean(total ** 2)

      MSE_list.append(MSE)

      R2 = 1 - (MSE / MST)
      R2_list.append(R2)

      param_0_list.append(param_0)
      param_1_list.append(param_1)

    return param_1_list, param_0_list, MSE_list, R2_list
```

```python
S = sgd(X, y, 0.001, 10)
# coefficient retrieval
param_1_list, param_0_list, MSE_list, R2_list = S[0], S[1], S[2], S[3]

fig, ax = plt.subplots(1,2,figsize=(15, 5))
ax1 = ax[0].twinx()

def update(i):
    ax[0].clear()
    ax[0].scatter(i, MSE_list[i], color='tab:red')
    ax[0].plot([0, i],[MSE_list[i], MSE_list[i]], color='tab:red',linestyle='--', label=f'MSE :{MSE_list[i]:.2f}')
    ax[0].plot([i, i],[0, MSE_list[i]], color='tab:green',linestyle='--')
    ax[0].plot(MSE_list, color='tab:red')
    ax[0].set_ylabel('MSE')
    ax[0].legend(loc='center right', bbox_to_anchor=(1, 0.5))

    ax1.clear()
    ax1.scatter(i, R2_list[i], color='tab:blue')
    ax1.plot([0, i],[R2_list[i], R2_list[i]], color='tab:blue',linestyle='--', label=f'R Square :{R2_list[i]:.2f}')
    ax1.plot([i, i],[0, R2_list[i]], color='tab:green', linestyle='--')
    ax1.plot(R2_list, color='tab:blue')
    ax1.set_ylabel('R Square')
    ax1.yaxis.set_label_position('right')
    ax1.legend(loc='center right', bbox_to_anchor=(1, 0.6))

    ax[0].set_xlabel('Iterations')
    ax[0].set_title(f'MSE & R Square vs Iteration\nIteration: {i}')

    ax[1].clear()
    ax[1].scatter(X, y, label='Data Points')
    # plotting linear regression line
    ax[1].plot(X, param_1_list[i] * X + param_0_list[i], color='tab:green', linestyle='--',label=f'Regression')
    ax[1].set_xlabel('X')
    ax[1].set_ylabel('y')
    ax[1].set_title(f'Simple Linear Regression using Stochastic Gradient Descent')
    ax[1].legend()

    plt.tight_layout()

ani = FuncAnimation(fig, update, frames=len(param_1_list), interval=100, repeat=False)

HTML(ani.to_jshtml())
```

```python
def mini_batch_sgd(X, y, param_1, param_0, learning_rate, num_iterations, batch_size):
  cost_list = []
  param_0_list = []
  param_1_list = []
  X_batches = []
  y_batches = []

  for j in range(0, len(X), batch_size):
    X_batch = X[j:j + batch_size]
    y_batch = y[j:j + batch_size]

    for i in range(num_iterations):
      y_pred = param_1 * X_batch + param_0
      loss = y_pred - y_batch

      cost = np.sum(loss**2) / (2 * len(loss))

      grad_0 = np.sum(loss) * (2 / len(loss))
      grad_1 = np.sum(loss * X_batch) * (2 / len(loss))

      param_0 -= learning_rate * grad_0
      param_1 -= learning_rate * grad_1

    cost_list.append(cost)
    param_0_list.append(param_0)
    param_1_list.append(param_1)
    X_batches.append(X_batch)
    y_batches.append(y_batch)

  return param_1_list, param_0_list, cost_list, X_batches, y_batches
```

```python
M = mini_batch_sgd(X_1, y_1, 0, 0, 0.1, 100, 10)
param_0_list, param_1_list, cost_list, X_batches, y_batches = M[1], M[0], M[2], M[3], M[4]

fig, ax = plt.subplots(1,2,figsize=(10, 6))
ax[0].plot(cost_list)
ax[0].set_xlabel('Iterations')
ax[0].set_ylabel('Cost')
ax[0].set_title('Gradient Descent Visualization')

def update(i):
    ax[1].clear()
    ax[1].scatter(X_1, y_1, color='blue', label='Data')
    ax[1].scatter(X_batches[i], y_batches[i], color='red', label=f'{i}th Batch')
    ax[1].plot(X_1, param_1_list[i] * X_1 + param_0_list[i], color='green',label=f'Regression - Cost : {cost_list[i]:.2f}')
    ax[1].set_xlabel('X')
    ax[1].set_ylabel('y')
    ax[1].set_title(f'Mini-Batch Stochastic Gradient Descent (Iteration {i})')
    ax[1].legend()

ani = FuncAnimation(fig, update, frames=len(param_1_list), interval=100, repeat=False)

HTML(ani.to_jshtml())
```
