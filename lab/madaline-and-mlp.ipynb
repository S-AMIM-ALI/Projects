# Madaline


```python
import numpy as np
```

```python
input_data = np.array([[0,0],
                     [0,1],
                     [1,0],
                     [1,1]
                      ])
target = np.array([0,0,0,1])
```

```python
def activation_function(x):
    return 1 if x>= 0 else 0
```

```python
#convert output to bipolar for madaline
target_bipolar = np.array([-1 if val == 0 else 1 for val in target])
print('target_bipolar ',target_bipolar)
w1= np.zeros(3)
w2 = np.zeros(3)
b1 = b2 = 0
epochs = 5
learning_rate = 0.1
```

```python
#Add bias term to input (as a columns of 1's)
input_date = np.hstack((input_data,np.ones((input_data.shape[0],1))))
```

```python
for epoch in range(epochs):
  for i in range(len(input_data)):
    input_sample = input_date[i]
    expected_sample = target_bipolar[i]

    z1_in = np.dot(input_sample,w1) + b1
    z2_in = np.dot(input_sample,w2) + b2
    z1 = activation_function(z1_in)
    z2 = activation_function(z2_in)
    output = activation_function(z1 + z2 -1)

    if output != expected_sample:
      #Minimum Adjustment : adjust one wrong neuron
      if z1 != expected_sample:
        w1 += learning_rate *input_sample*expected_sample
        b1 += learning_rate * expected_sample
      if z2 != expected_sample:
         w2 += learning_rate *input_sample*expected_sample
         b2 += learning_rate * expected_sample

print(f"Final weights w1: {w1}, w2: {w2}")
print(f"Final bias b1: {b1}, b2: {b2}")
```
